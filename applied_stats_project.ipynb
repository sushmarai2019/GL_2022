```python
# 2.A. Probability that none of the items are defective? [1 Mark]

n=10  # 10 manufactured items is selected
p=.05  # failure rate for the manufactured item is 5%
k=np.arange(0,11)
k   # 10 manufactured items is selected
binomial = stats.binom.pmf(k,n,p)  # probabilty mass function
print(binomial[0])
print('2.A.There is:',binomial[0],'chance that none of the items are defective')
```

```python
# 2.B. Probability that exactly one of the items is defective? [1 Mark]

print(binomial[1])
print('2.B. There is:',binomial[1],'chance that one of the items is defective')
```

```python
# 2.C. Probabiliity that two or fewer of the items are defectve? [1 Mark]
#p(X=0) + p(X=1) + p(X=2) 
print(binomial[0]+binomial[1]+binomial[2])
print('2.C.There is:',binomial[0]+binomial[1]+binomial[2],'two or fewer of the items are defectve')
```

```python
# 2.D. Probability that three or more of the items are defective
#p(X>=3)
for i in binomial[3:11]:
    val=0
    val=val+i
print('2.D.There is:',val,'chance that three or more of the items are defective')
```

```python
# 3.A. What is Probability that in a given week he will sell some cars? [1 Mark]
# X >= 1 {1 or more cars}
#Lamda = 3 {average car sold per week}
#1-p[X<1] => 1-p[X=0]

rate = 3
#a numpy array of different no. of cars sold per week

n=np.arange(0,10)

#calculating the distribution and storing the distribution of probablitites in an array
poisson1 = stats.poisson.pmf(n,rate)
print('3.A.There is:',1-poisson1[0],'chance that in a given week he will sell some cars')
```

```python
#3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars? [1 Mark]
print('3.B.There is:',poisson1[2]+poisson1[3]+poisson1[4],'chance that in a given week he will sell 2 or more but less than 5 cars')
```

```python
#3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week. [1 Mark]
#visualising the plot with pyplot
cdf = poisson.cdf(n,mu=rate)
plt.plot(n,cdf,'o-')
plt.title('Poisson: $\lambda$ = %i ' % rate)
plt.xlabel('cars sold per-week')
plt.ylabel('cumulative probability of cars sold per-week')
plt.show()
```

```python
# 4.A. What is the probability that all three orders will be recognised correctly? [1 Mark]
n=3  # 3 orders placed independently
p=.868  # Recognition accuracy that measures the percentage of orders that are taken correctly
k=np.arange(0,11)
k   # 10 manufactured items is selected
binomial_orders = stats.binom.pmf(k,n,p)  # probabilty mass function
print(binomial_orders[3])
print('4.A.There is:',binomial_orders[3],'chance that all three orders will be recognised correctly')
```

```python
# 4.B. What is the probability that none of the three orders will be recognised correctly? [1 Mark]
print(binomial_orders[0])
print('4.B.There is:',binomial_orders[0],'chance that none of the three orders will be recognised correctly')
```

```python
# 4.C. What is the probability that at least two of the three orders will be recognised correctly? [1 Mark]
print(binomial_orders[2]+binomial_orders[3])
print('4.C.There is:',binomial_orders[2]+binomial_orders[3],'chance that at least two of the three orders will be recognised correctly')
```

```python
#Part C - 15 Marks
#• DOMAIN: Startup ecosystem
#• CONTEXT: Company X is a EU online publisher focusing on the startups industry. The company specifically reports on the business related to
#technology news, analysis of emerging trends and profiling of new tech businesses and products. Their event i.e. Startup Battlefield is the world’s
#pre-eminent startup competition. Startup Battlefield features 15-30 top early stage startups pitching top judges in front of a vast live audience,
#present in person and online.
#• DATA DESCRIPTION: CompanyX_EU.csv - Each row in the dataset is a Start-up company and the columns describe the company.
#• DATA DICTIONARY:
#1. Startup: Name of the company
#2. Product: Actual product
#3. Funding: Funds raised by the company in USD
#4. Event: The event the company participated in
#5. Result: Described by Contestant, Finalist, Audience choice, Winner or Runner up
#6. OperatingState: Current status of the company, Operating ,Closed, Acquired or IPO
#*Dataset has been downloaded from the internet. All the credit for the dataset goes to the original creator of the data.
#• PROJECT OBJECTIVE: Analyse the data of the various companies from the given dataset and perform the tasks that are specified in the below
#steps. Draw insights from the various attributes that are present in the dataset, plot distributions, state hypotheses and draw conclusions from
#the dataset.
```

```python
# Read the Data 
df_cmp= pd.read_csv('CompanyX_EU.csv')
df_cmp.head(8) # View first n rows of data 
```

```python
#2A. Check the datatypes of each attribute.
df_cmp.info()  # list of columns and object info
print(df_cmp.dtypes)
```

```python
#2B. Check for null values in the attributes.
df_cmp.count()
```

```python
#3.A Drop the null values. [1 Mark]
df_dropped = df_cmp.dropna()  # Dont do this on the original dataframe, you will lose good rows with values , make a copy
df_dropped=df_dropped.reset_index(drop=True)
#Get the shape of the dataframe after removing the null values
df_dropped.shape
```

```python
#3.B. Convert the ‘Funding’ features to a numerical value
df_dropped.loc[:,'Funds_in_million'] = df_dropped['Funding'].apply(lambda x: float(x[1:-1])/1000 if x[-1] == 'K' else (float(x[1:-1])*1000 if x[-1] == 'B' else float(x[1:-1])))
print(df_dropped.dtypes)
print(df_dropped.head(5))
```

```python
#3 C. Plot box plot for funds in million. [1 Mark]
plot = plt.boxplot(df_dropped.Funds_in_million)
plt.title('Boxplot of the funds')
plt.ylabel("Funds raised (in Million)")

plt.show()
```

```python
upper_fence = plot['caps'][1].get_data()[1][1]   # we can use the values from the box plot itself to get the upper fence
upper_fence
```

```python
#also checking the distribution plot
sns.distplot(df_dropped["Funds_in_million"]);
```

```python
#3 D. Check the number of outliers greater than the upper fence. [1 Mark]
q1= df_dropped["Funds_in_million"].quantile(0.25)
q3= df_dropped["Funds_in_million"].quantile(0.75)
iqr = q3-q1
#upper_fence = 3*iqr
df_dropped_outlier = []
for index, x in enumerate(df_dropped["Funds_in_million"]):
    if x >= upper_fence:
       df_dropped_outlier.append(index)
print(df_dropped_outlier)    
print('Count of possible outliers are',len(df_dropped_outlier) )
```

```python
#droppping the outliers
df_dropped.drop(df_dropped[df_dropped.Funds_in_million > upper_fence].index, inplace = True)
plot = plt.boxplot(df_dropped.Funds_in_million)
plt.title('Boxplot of the funds')
plt.ylabel("Funds raised (in Million)")

plt.show()
```

```python
#3 E. Check frequency of the OperatingState features classes. [1 Mark]
sns.countplot(df_dropped["OperatingState"]);
```

```python
#4.A. Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down? [1 Mark]

#fig, ax = plt.subplots(figsize=(20,10))
options = ['Closed','Operating']
df_graph = df_dropped[df_dropped['OperatingState'].isin(options)]
sns.catplot('OperatingState','Funds_in_million', jitter=False, hue="OperatingState",data=df_graph)
print('after dropping outliers We are not fully able to draw this conclusion that companies still operating and raising more funds that companies which are closed')
```

```python
#4B. Write the null hypothesis and alternative hypothesis. [1 Mark]
print("As we are not able to conclude that operating companies are raising more fund as comapred to closed companies hence we have to compare the means")
print("Null hypothesis (Ho) : There is no difference between the two means")
print("Alternate hypothesis (Ha) : There is significant difference between the two means")
```

```python
#4C Test for significance and conclusion [1 Mark]
from statsmodels.stats.weightstats import ztest

sample1 = df_dropped.loc[df_dropped.OperatingState == 'Operating', 'Funds_in_million']
sample2 = df_dropped.loc[df_dropped.OperatingState =='Closed', 'Funds_in_million']

alpha = 0.05   # Let's consider a significance level of 5%

test_statistic, p_value = ztest(sample1, sample2)

if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
```

```python
#4D . Make a copy of the original data frame. [1 Mark]
# Read the Data 
df_cmp_copy= df_cmp.copy(deep = True)
df_cmp_copy.head(8) # View first n rows of data 
```

```python
#4E. Check frequency distribution of Result variables. [1 Mark]
#also checking the distribution plot
catergories = df_cmp_copy["Result"].value_counts().index
counts = df_cmp_copy["Result"].value_counts().values
plt.bar(catergories,counts,width=0.5)
plt.xlabel("Result column catergories")
plt.ylabel("Result column count")
```

```python
#4F Calculate percentage of winners that are still operating and percentage of contestants that are still operating [1 Mark]
#find the columns count where winner and operating 
winner_operating = df_cmp_copy[(df_cmp_copy.Result=='Winner') &(df_cmp_copy.OperatingState=='Operating')].count().loc['Result']
#find the total count where winner and operating 
winner=df_cmp_copy[df_cmp_copy['Result']=='Winner'].count().loc['Result']

#find the columns count where contestants and operating 
contestants_operating = df_cmp_copy[(df_cmp_copy.Result=='Contestant') &(df_cmp_copy.OperatingState=='Operating')].count().loc['Result']
#find the total count where winner and operating 
contestants=df_cmp_copy[df_cmp_copy['Result']=='Contestant'].count().loc['Result']

print('percentage of winners who are still operating :',winner_operating/winner)
print('percentage of contestants who are still operating :',contestants_operating/contestants)
```

```python
#4G. Write your hypothesis comparing the proportion of companies that are operating between winners and contestants: [2 Mark]
#since its between winners and contestants hecne 2 tails test is required
print('Null hyputhesis (Ho): The proportion of companies that are operating is the same in both categories - winners and contestants')
print('Alternative hypothesis (Ha): The proportion of companies that are operating is different from each other, among the two categories')
```

```python
#4H. Test for significance and conclusion [1 Mark]
from statsmodels.stats.proportion import proportions_ztest

test_statistic, p_value = proportions_ztest([contestants_operating, winner_operating], [contestants, winner])
alpha = 0.05   # Let's consider a significance level of 5%

if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')

print('The amount of funds a company raises is not a significant factor in deciding if the company will remain operational or close down')
print('Winners in the tech-crunch battle have proven to remain operational more so, than the companies that came out only as contestants Bonus')
```
